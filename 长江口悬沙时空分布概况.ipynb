水动力是河口地貌演变的主要动力，而泥沙则是其物质基础。长江口悬浮泥沙从徐六泾到口门再到口外，悬沙浓度不断升高*陈沈良*，口外部分泥沙含量从南向北部显示出中间高而两边低的形态，南部略高于北部，与河口汊道方向基本一致，最高浓度的区域出现在拦门沙前缘浑水峰水域。泥沙扩散形状呈羽毛状和喷射状，喷射状主要是由于浑浊水体径流在河槽的约束下受到落潮动力轴惯性作用的影响，而河口冲淡水在盐水楔之上，漂浮于海绵，在风浪作用下运动方向容易发生改变，在扩散前沿形成羽毛状分布*云才行*。最大浑浊带并非长江口悬沙浓度最高的区域。长江口悬沙浓度存在明显的季节和大小潮周期差异，在长江口徐六泾水域，大潮流速远大于小潮，悬沙浓度大潮大于小潮，并且泥沙浓度峰滞后于流速峰，落潮悬沙浓度大于涨潮*程江*。随着潮汐作用的加强，从口内到口外，大小潮周期悬沙浓度的震荡幅度不断增加，与徐六泾相同，出现大潮时再悬浮作用下的高浓度特征*陈沈良*。口内夏季大于冬季，口外冬季大于夏季，主要原因是夏季径流作用强，而海洋动力作用相对较弱，造成泥沙再悬浮作用减弱，而冬季径流作用弱，潮汐和风浪等海洋作用强，泥沙再悬浮率明显高于夏季，造成口门附近的洪冲枯淤的结果。在年季变化上，从1974年开始到1989年，悬沙浓度呈缓慢上升势态，1989年到1997年处于快速下降时期，1997年相比1989年下降了约40%，从1997年到2002年呈波动状态，2002年以后北支略有上涨，而南支依然处于波动平衡状态，2006年入海泥沙量为历史最低值*陈勇*。
